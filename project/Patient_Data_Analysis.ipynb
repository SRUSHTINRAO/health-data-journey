```python
import pandas as pd
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [23, 45, 30, 25, 40],
    "Disease": ["Diabetes", "BP", "Diabetes", "Normal", "BP"]
}

df = pd.DataFrame(data)
```

```python
# Average age
df["Age"].mean()

```

```python
# Count diseases
df["Disease"].value_counts()
```

```python
# Filter high risk
df[df["Age"] > 30]
```

```python
import matplotlib.pyplot as plt
df["Disease"].value_counts().plot(kind="bar")
plt.show()
```

```python
df.groupby("Disease")["Age"].mean()
```
